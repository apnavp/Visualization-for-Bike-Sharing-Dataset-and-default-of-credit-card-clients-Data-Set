# II. Linear Algebra and Probability Theory


## A. Linear Algebra

   Understanding of Linear Algebra is important to understand the machine learning algorithms, this includes study of mathematicals objects such as scalars,vectors,matrices,tensors. Linear Algebra provides many operations like multipy on vectors and matrices, Norm which is a function to measure the size of vector, Decomposition of a matrix to understand it better by Eigendecomposition:
    
   \begin{equation*}
    \left( A = Vdiag(\lambda) V^{-1} \right)
    \end{equation*}
        where V is the matrix of all the eigen vectors,
          $ \lambda$ is the matrix of all eigen values.
         
 The trace operator which gives the sum of all diagonal entries of a matrix by:
   
   \begin{equation*}
    \left( Tr(A) = \sum_{i} a_i,_i \right)
    \end{equation*}

Principal Component Analysis is one of the machine learning algorithms that can be derived only using linear algebra.

## B. Probability Theory

Probability Theory in AI is used in two major ways: first the law of probability tell us how AI should reason, second we can use probability and statistics to theoretically analyze the behavior of proposed AI.Machine learning mostly deals with uncertain quantities and probability theory is a mathematical framework to represent uncertain statements.
Probability distribution can be shown as how likely a random variable is to take on each of its possible states.The way to describe probability distribution depends upon the variable being discrete or continuos. Commom probability distributions include bernoulli distribution,multinoulli distribution,gaussian distribution. Probability theory consists of bayes rule, where if we know the P(x|y) and P(x) but want to find P(y|x) we can compute it using staright forward formula known as bayes rule:

   \begin{equation*}
    \left( P(x|y) = \frac{P(x) P(y|x)}{P(y)} \right)
    \end{equation*}

weekday = day of the week 
<br>
workingday = if day is neither weekend nor holiday is 1, otherwise is 0. 
<br> 
weathersit :
<br>
1: Clear, Few clouds, Partly cloudy, Partly cloudy 
<br>
2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist

3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds 
<br>
4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog 
<br>
temp = Normalized temperature in Celsius. The values are derived via
    \begin{equation*}
    \frac{t-t_{min}}{t_{max}-t_{min}} \\
    t_{min}=-8, t_{max}=+39

atemp= Normalized feeling temperature in Celsius. The values are derived via 
    \begin{equation*}
    \frac{t-t_{min}}{t_{max}-t_{min}} \\

    \end{equation*}

<br>
hum= Normalized humidity. The values are divided to 100 (max) 
<br>
windspeed: Normalized wind speed. The values are divided to 67 (max) 
<br>
casual= count of casual users 
<br>
registered: count of registered users 
<br>
cnt= count of total rental bikes including both casual and registered 
<br>

## D.Visualization of Data and Preliminary Observations

```python
plt.rcParams["figure.figsize"] = (10,10)
x = [1,2,3,4]
labels = ['Winter', 'Summer', 'Spring','Fall']
plt.xticks(x, labels)
plt.xlabel('Seasons', fontsize='large')
plt.ylabel('Total Bike Rentals', fontsize='large');
plt.scatter('season', 'cnt', data=dataset, c='season')
```

This scatter plot shows the count of total bikes rented in that season with respect to windspeed. It can be clearly seen that bike rentals are inversely proportional with windspeed.

### Weekdays vs Total Bike Rentals

### Humidity vs Total Bike Rentals

### Hour of day vs Total Bike Rentals

## B.Reading the Data

Reading the .csv file for processing the data.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

dataset1=pd.read_excel('default of credit card clients.xls')
```

## C.Preprocessing the Data

Checking if dataset has any null values.

```python
dataset1.isnull().sum()
```

As there are no null values there is no need for preprocessing.

```python
dataset1.describe()
```

## D.Visualization of Data and  Preliminary Observations

```python
plt.rcParams["figure.figsize"] = (8,5)
sns.countplot(dataset1['default payment next month'],label="Count").set_xticklabels(["No", "Yes"])
plt.show()
```

### Ratio of default payment next month
This graph shows the count of sex with respect to default payment next month. It can be clearly seen that the graph for less number of users will be paying credit card payments on time.

```python
sns.catplot(x='SEX',y='default payment next month',data=dataset1,kind='bar'
).set_axis_labels("Sex","default payment next month").set_xticklabels(["Men", "Women"])
plt.show()
```

### Sex vs default payment next month
This catplot shows the Gender ratio with default payment method. It can be seen that males have higher on-time payment ratio of females.


```python
sns.catplot(x='EDUCATION',y='default payment next month',data=dataset1,kind='bar').set_axis_labels("Education","default payment next month").set_xticklabels(["Graduate school","University", "High school" , "Others"])
plt.show()
```

### Education vs default payment next month
This catplot shows the effect of education on default payment next month, as it can be seen that with increase in education the default payment next month probability of the user also increases.

```python
sns.catplot(x='MARRIAGE',y='default payment next month',data=dataset1,kind='bar').set_axis_labels("Marriage","default payment next month").set_xticklabels(["Married","Single", "Others" ])
plt.show()
```

### Marriage vs default payment next month
This catplot shows the effect of marital status of an adult on his/her default payment next month,here it can be seen that adults who are married have greater probability of default payment next month.

```python
sns.catplot(x='AGE',y='default payment next month',data=dataset1,kind='bar',height=11).set_axis_labels("Age(year)","default payment next month")
plt.show()
```

### Age vs default payment next month
This catplot shows the effect of age of an adult on his/her default payment next month,here it can be seen that adults who are age 60-61 have greater probability of default payment next month.

## E.Preliminary observation
From the above catplots we can conclude that users which are male, have education higher than high school, married and age in peak range 60 and 61 have highest probablity of paying credit card bills on time.
